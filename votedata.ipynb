```python
import pandas as pd
df = pd.read_excel('600(kaohsuing).xls',header = 1)
```

```python
data_length = df.shape[0]
data_width = df.shape[1]
print(data_length,data_width)
df.dropna(how='all',inplace=True)
df.drop(columns=['H\n投票率\nH=C/G\n(%)','投開票所別'],inplace=True)

for i in range(2,df.shape[0]):
    if df.iloc[i,1] == df.iloc[i-1,1]:
        for j in range(3,df.shape[1]):
            df.iloc[i,j] = df.iloc[i,j] + df.iloc[i-1,j]
for i in range(df.shape[0]):
    if df.iloc[i,0] != "　　":
        df.iloc[i,1] = str(df.iloc[i,0]) + "合計"
df.行政區別.replace("　",None,  inplace=True)
df.行政區別.replace("　　",None,  inplace=True)

df.drop_duplicates(subset=['村里別'],keep='last',inplace=True)
df.set_index('行政區別')
df.to_excel('TEST_result.xlsx',sheet_name='sheet1')
```
